```python
import pickle
import copy
from network_portrait_divergence_master import portrait_divergence as pd
diffVec1=dict()
diffVec2=dict()
clustVec1=dict()
clustVec2=dict()
clustVec3=dict()
transVec1=dict()
transVec2=dict()
transVec3=dict()
avDegree1=dict()
avDegree2=dict()
avDegree3=dict()
hierarchy1=dict()
hierarchy2=dict()
hierarchy3=dict()
modelDiffToArabVec=dict()
arabDiffToArabVec=dict()
netVec1=dict()
netVec2=dict()
netVec3=dict()
G_seed1=nx.DiGraph()
G_seed1.add_edge(0,0)


#with open('E_coli_GRN/seed_for_comparisons_of_measures.pkl', 'wb') as f:
#    pickle.dump(G_seed2, f)

#p=anc.fit_p_degree_recurrence(G_coli,G_seed2,0.000001)
steps=5000
qRange=[0.3,0.4,0.5,0.6,0.7,0.8]
sampleNo=50
samplePoints=[0,100,200,500,1000,2000,5000]
G_seed2=nx.DiGraph(ut.met_hastings_connected_induced_subgraph(G_arab,100,5000,10))
G_seed3=nx.DiGraph(ut.met_hastings_connected_induced_subgraph(G_coli,100,5000,10))
for q in qRange:
    diffVec1[q]=dict()
    diffVec2[q]=dict()
    clustVec1[q]=dict()
    clustVec2[q]=dict()
    clustVec3[q]=dict()
    transVec1[q]=dict()
    transVec2[q]=dict()
    transVec3[q]=dict()
    avDegree1[q]=dict()
    avDegree2[q]=dict()
    avDegree3[q]=dict()
    hierarchy1[q]=dict()
    hierarchy2[q]=dict()
    hierarchy3[q]=dict()
    netVec1[q]=dict()
    netVec2[q]=dict()
    netVec3[q]=dict()
    modelDiffToArabVec[q]=[]
    arabDiffToArabVec[q]=[]
    for i in range(sampleNo):
        diffVec1[q][i]=[]
        diffVec2[q][i]=[]
        clustVec1[q][i]=[]
        clustVec2[q][i]=[]
        clustVec3[q][i]=[]
        transVec1[q][i]=[]
        transVec2[q][i]=[]
        transVec3[q][i]=[]
        avDegree1[q][i]=[]
        avDegree2[q][i]=[]
        avDegree3[q][i]=[]
        hierarchy1[q][i]=[]
        hierarchy2[q][i]=[]
        hierarchy3[q][i]=[]
        netVec1[q][i]=[]
        netVec2[q][i]=[]
        netVec3[q][i]=[]
        seed_size=len(G_seed2.nodes)
        G1=copy.deepcopy(G_seed1)
        G2=copy.deepcopy(G_seed2)
        G3=copy.deepcopy(G_seed3)

        m1=0
        m2=0
        m3=0
        while m1<seed_size:
            G1=dup.PED_PEA(G1,0,q,m1+1,copyG=False)
            node_remover=list(nx.isolates(G1))
            if node_remover==[]:
                m1=m1+1
            else:
                G1.remove_nodes_from(node_remover)

        print(m1)    
        #diffVec1[q][i].append(pd.portrait_divergence(G1,G2))
        #diffVec2[q][i].append(pd.portrait_divergence(G1,G3))
        #clustVec1[q][i].append(nx.average_clustering(G1))
        #clustVec2[q][i].append(nx.average_clustering(G2))
        #clustVec3[q][i].append(nx.average_clustering(G3))
        transVec1[q][i].append(nx.transitivity(G1))
        transVec2[q][i].append(nx.transitivity(G2))
        transVec3[q][i].append(nx.transitivity(G3))
        avDegree1[q][i].append(sum([d for (n, d) in nx.degree(G1)]) / len(G1.nodes()))
        avDegree2[q][i].append(sum([d for (n, d) in nx.degree(G2)]) / len(G2.nodes()))
        avDegree3[q][i].append(sum([d for (n, d) in nx.degree(G3)]) / len(G3.nodes()))
        hierarchy1[q][i].append(nx.global_reaching_centrality(G1))
        hierarchy2[q][i].append(nx.global_reaching_centrality(G2))
        hierarchy3[q][i].append(nx.global_reaching_centrality(G3))
        netVec1[q][i].append(G1)
        netVec2[q][i].append(G2)
        netVec3[q][i].append(G3)
        while m1<steps:
                
                G1=dup.PED_PEA(G1,0,q,m1+1,copyG=False)
                node_remover=list(nx.isolates(G1))
                if node_remover==[]:
                    m1=m1+1
                    trigger1=False
                    while not trigger1:
                        G2=dup.PED_PEA(G2,0,q,m2+1,copyG=False)
                        node_remover=list(nx.isolates(G2))
                        if node_remover==[]:
                            trigger1=True
                            m2=m2+1
                            trigger2=False
                            while not trigger2:
                                G3=dup.PED_PEA(G3,0,q,m3+1,copyG=False)
                                node_remover=list(nx.isolates(G3))
                                if node_remover==[]:
                                    trigger2=True
                                    m3=m3+1
                                    if m1 in samplePoints:
                                        
                                        print(i,m1,m2,m3)

                                        #diffVec1[q][i].append(pd.portrait_divergence(G1,G2))
                                        #diffVec2[q][i].append(pd.portrait_divergence(G1,G3))
                                        #clustVec1[q][i].append(nx.average_clustering(G1))
                                        #clustVec2[q][i].append(nx.average_clustering(G2))
                                        #clustVec3[q][i].append(nx.average_clustering(G3))
                                        transVec1[q][i].append(nx.transitivity(G1))
                                        transVec2[q][i].append(nx.transitivity(G2))
                                        transVec3[q][i].append(nx.transitivity(G3))
                                        avDegree1[q][i].append(sum([d for (n, d) in nx.degree(G1)]) / len(G1.nodes()))
                                        avDegree2[q][i].append(sum([d for (n, d) in nx.degree(G2)]) / len(G2.nodes()))
                                        avDegree3[q][i].append(sum([d for (n, d) in nx.degree(G3)]) / len(G3.nodes()))
                                        hierarchy1[q][i].append(nx.global_reaching_centrality(G1))
                                        hierarchy2[q][i].append(nx.global_reaching_centrality(G2))
                                        hierarchy3[q][i].append(nx.global_reaching_centrality(G3))
                                        netVec1[q][i].append(G1)
                                        netVec2[q][i].append(G2)
                                        netVec3[q][i].append(G3)
                                else:
                                    G3.remove_nodes_from(node_remover)
                        
                        else:
                            G2.remove_nodes_from(node_remover)
                else:
                    
                    G1.remove_nodes_from(node_remover)
        #modelDiffToArabVec[q].append(pd.portrait_divergence(G1,G_arab))
        #rabDiffToArabVec[q].append(pd.portrait_divergence(G2,G_arab))
        #modelDiffToColiVec[q].append(pd.portrait_divergence(G1,G_coli))
        #coliDiffToColiVec[q].append(pd.portrait_divergence(G3,G_coli))
with open('arabidopsis_GRN/model_arabidopsis_diffVec1.pkl', 'wb') as f:
    pickle.dump(diffVec1, f)
with open('arabidopsis_GRN/model_arabidopsis_diffVec2.pkl', 'wb') as f:
    pickle.dump(diffVec2, f)
with open('arabidopsis_GRN/model_arabidopsis_clustVec1.pkl', 'wb') as f:
    pickle.dump(clustVec1, f)
with open('arabidopsis_GRN/model_arabidopsis_clustVec2.pkl', 'wb') as f:
    pickle.dump(clustVec2, f)
with open('arabidopsis_GRN/model_arabidopsis_clustVec3.pkl', 'wb') as f:
    pickle.dump(clustVec3, f)
with open('arabidopsis_GRN/model_arabidopsis_transVec1.pkl', 'wb') as f:
    pickle.dump(transVec1, f)
with open('arabidopsis_GRN/model_arabidopsis_transVec2.pkl', 'wb') as f:
    pickle.dump(transVec2, f)
with open('arabidopsis_GRN/model_arabidopsis_transVec3.pkl', 'wb') as f:
    pickle.dump(transVec3, f)
with open('arabidopsis_GRN/model_arabidopsis_avDegree1.pkl', 'wb') as f:
    pickle.dump(avDegree1, f)
with open('arabidopsis_GRN/model_arabidopsis_avDegree2.pkl', 'wb') as f:
    pickle.dump(avDegree2, f)
with open('arabidopsis_GRN/model_arabidopsis_avDegree3.pkl', 'wb') as f:
    pickle.dump(avDegree3, f)
with open('arabidopsis_GRN/model_arabidopsis_hierarchy1.pkl','wb') as f:
    pickle.dump(hierarchy1,f)
with open('arabidopsis_GRN/model_arabidopsis_hierarchy2.pkl','wb') as f:
    pickle.dump(hierarchy2,f)
with open('arabidopsis_GRN/model_arabidopsis_hierarchy3.pkl','wb') as f:
    pickle.dump(hierarchy3,f)
with open('arabidopsis_GRN/model_arabidopsis_netVec1.pkl', 'wb') as f:
    pickle.dump(netVec1, f)
with open('arabidopsis_GRN/model_arabidopsis_netVec2.pkl', 'wb') as f:
    pickle.dump(netVec2, f)
with open('arabidopsis_GRN/model_arabidopsis_netVec3.pkl', 'wb') as f:
    pickle.dump(netVec3, f)        
```
